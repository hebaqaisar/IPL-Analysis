```python
#To import all the libraries that would be used for this project 

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#To extract the data from the file

ipl =  pd.read_csv("C:\\Users\\HIBA QAISAR\\Downloads\\data.csv")
```

```python
#To read the data 

ipl.head()
```

```python
#To count the total number of rows and coloums (no, of rows, no. of cols)

ipl.shape
```

```python
#To count total number of times the top 5 players won the 'player of the match' award 

ipl['player_of_match'].value_counts()[0:5]
```

```python
#To count the number of times a team has won a match

ipl['winner'].value_counts()[:5]
```

```python
#To count the number of matches played each year

ipl['season'].value_counts()
```

```python
#barplot to get total number of times the top 5 players won the 'player of the match' award 

plt.figure(figsize = (8,4))
plt.bar(ipl['player_of_match'].value_counts().keys()[:5], ipl['player_of_match'].value_counts()[:5], color = 'green')
```

```python
#To check how many matches resulted in a tie, normal(lose or win) or no result(match didnt take place due to some unforeseen reasons)

ipl['result'].value_counts()
```

```python
#To extract the teams that won match by batting first

batting_first = ipl[ipl['win_by_runs']!=0]
```

```python
#To view the table

batting_first.head()
```

```python
#To extract the top five teams who won by batting first.

batting_first['winner'].keys()[:5]
```

```python
#bar graph to show the top five teams who won by batting first.

plt.figure(figsize = (7,4))

plt.bar(batting_first['winner'].value_counts().keys()[:3], batting_first['winner'].value_counts()[:3], color=['blue','purple','green'])
```

```python
#pie chart to show the top five teams who won by batting first.

plt.figure(figsize = (8,8))
plt.pie(batting_first['winner'].value_counts(),labels= batting_first['winner'].value_counts().keys() , autopct= '%0.1f%%')
```
